```python
import openml
from gplearn.genetic import SymbolicTransformer, SymbolicRegressor
from evolutionary_forest.forest import EvolutionaryForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
dataset = openml.datasets.get_dataset(547)

X, y, _, _ = dataset.get_data(dataset_format="dataframe", target=dataset.default_target_attribute)
```

```python
X = (X - X.mean()) / X.std()
```

```python
y = (y - y.mean()) / y.std()
```

```python
transformer = SymbolicTransformer()
transformer.fit(X, y)
```

```python
X_transformed = transformer.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
r = EvolutionaryForestRegressor()
r.fit(X_train, y_train)
print("R2 Score of test samples whit the EF model: ", r2_score(y_test, r.predict(X_test)))
```

```python
est_gp = SymbolicRegressor()
est_gp.fit(X_train, y_train)
print("R2 Score of test samples whit the SRGP model: ", r2_score(y_test, est_gp.predict(X_test)))
```
